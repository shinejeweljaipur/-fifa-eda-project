```python
# =============================================================
# FIFA WORLD CUP 2026 - FULL RELATIONAL DATASET EDA (NumPy + Pandas)
# 12 tables: matches, teams, players, venues, referees, events, etc.
# =============================================================

import numpy as np
import pandas as pd

pd.set_option('display.max_columns', 30)
pd.set_option('display.width', 150)

# -------------------------------------------------------------
# STEP 0 : LOAD ALL RAW TABLES
# -------------------------------------------------------------
matches            = pd.read_csv('matches.csv')
matches_detailed   = pd.read_csv('matches_detailed.csv')
match_events        = pd.read_csv('match_events.csv')
match_lineups       = pd.read_csv('match_lineups.csv')
match_team_stats    = pd.read_csv('match_team_stats.csv')
match_pred_features = pd.read_csv('match_prediction_features.csv')
player_stats        = pd.read_csv('player_stats.csv')
referees            = pd.read_csv('referees.csv')
squads_and_players   = pd.read_csv('squads_and_players.csv')
teams               = pd.read_csv('teams.csv')
tournament_stages    = pd.read_csv('tournament_stages.csv')
venues              = pd.read_csv('venues.csv')

tables = {
    'matches': matches, 'matches_detailed': matches_detailed, 'match_events': match_events,
    'match_lineups': match_lineups, 'match_team_stats': match_team_stats,
    'match_prediction_features': match_pred_features, 'player_stats': player_stats,
    'referees': referees, 'squads_and_players': squads_and_players, 'teams': teams,
    'tournament_stages': tournament_stages, 'venues': venues
}

# -------------------------------------------------------------
# STEP 1 : BASIC OVERVIEW OF EACH TABLE
# -------------------------------------------------------------
for name, df in tables.items():
    print(f"\n--- {name} ---")
    print("Shape:", df.shape)
    print("Columns:", df.columns.tolist())
    print("Dtypes:\n", df.dtypes)

# -------------------------------------------------------------
# STEP 2 : MISSING VALUES ACROSS ALL TABLES
# -------------------------------------------------------------
print("\n=== MISSING VALUES SUMMARY (per table) ===")
for name, df in tables.items():
    miss = df.isnull().sum()
    miss = miss[miss > 0]
    if len(miss) > 0:
        print(f"\n{name}:\n", miss)
        print("% missing:\n", (miss / len(df) * 100).round(2))

# -------------------------------------------------------------
# STEP 3 : DUPLICATES CHECK
# -------------------------------------------------------------
print("\n=== DUPLICATE ROWS PER TABLE ===")
for name, df in tables.items():
    print(f"{name}: {df.duplicated().sum()} duplicate rows")

# -------------------------------------------------------------
# STEP 4 : HANDLING MISSING VALUES (sensible per-column logic)
# -------------------------------------------------------------
# matches.csv: penalty scores are mostly NaN because most matches had no shootout -> fill with 0
matches['home_penalty_score'] = matches['home_penalty_score'].fillna(0)
matches['away_penalty_score'] = matches['away_penalty_score'].fillna(0)

# Matches not yet played (scores/xg missing) -> keep as NaN but flag them
matches['is_played'] = matches['home_score'].notnull()
print("\nMatches not yet played:", (~matches['is_played']).sum())

# player_stats.csv: goalkeeper-only fields (shots, rating) missing for outfield vs GK differently -> fill 0 where sensible
for col in ['shots', 'shots_on_target', 'clean_sheets', 'saves', 'goals_conceded']:
    player_stats[col] = player_stats[col].fillna(0)

# Flag columns that are ENTIRELY empty (100% missing) -> mean/median would also be NaN, so fill with 0
# and record them separately so they aren't silently mistaken for real zeros downstream.
fully_empty_cols = [c for c in player_stats.columns if player_stats[c].isnull().sum() == len(player_stats)]
print("\nColumns that are FULLY empty in player_stats (data quality issue):", fully_empty_cols)

if 'average_rating' in fully_empty_cols:
    player_stats['average_rating'] = player_stats['average_rating'].fillna(0)   # no real values exist to average
else:
    player_stats['average_rating'] = player_stats['average_rating'].fillna(player_stats['average_rating'].mean())

# match_team_stats: player_of_the_match missing for the losing/other team row -> fill 'None'
match_team_stats['player_of_the_match'] = match_team_stats['player_of_the_match'].fillna('None')

print("\nMissing after cleaning - matches:", matches.isnull().sum().sum() - matches[['home_score','away_score','result_type','home_xg','away_xg','player_of_the_match_id','home_team_id','away_team_id']].isnull().sum().sum())
print("Missing after cleaning - player_stats:", player_stats.isnull().sum().sum())

# -------------------------------------------------------------
# STEP 5 : DATA TYPE CONVERSIONS
# -------------------------------------------------------------
matches['date'] = pd.to_datetime(matches['date'], errors='coerce')
matches_detailed['date'] = pd.to_datetime(matches_detailed['date'], errors='coerce')
squads_and_players['date_of_birth'] = pd.to_datetime(squads_and_players['date_of_birth'], errors='coerce')

# -------------------------------------------------------------
# STEP 6 : STATISTICAL SUMMARY (NumPy + Pandas)
# -------------------------------------------------------------
print("\nTeams - numeric summary:\n", teams.describe())
print("\nPlayer stats - numeric summary:\n", player_stats.describe())

print("\nMean squad market value:", np.mean(squads_and_players['market_value_eur']))
print("Median caps:", np.median(squads_and_players['caps']))
print("Std of player height:", np.std(squads_and_players['height_cm']))
print("Percentiles of market value (25/50/75/90):",
      np.percentile(squads_and_players['market_value_eur'], [25, 50, 75, 90]))

# -------------------------------------------------------------
# STEP 6B : STANDARD DEVIATION (population vs sample) - NumPy & Pandas
# -------------------------------------------------------------
# NumPy's np.std() defaults to POPULATION std (ddof=0).
# Pandas' .std() defaults to SAMPLE std (ddof=1). Both are shown so the difference is clear.

std_cols = ['market_value_eur', 'caps', 'height_cm', 'goals']
print("\n--- Standard Deviation: squads_and_players ---")
for col in std_cols:
    pop_std = np.std(squads_and_players[col])            # ddof=0 (population)
    sample_std = squads_and_players[col].std()            # ddof=1 (sample) - pandas default
    print(f"{col}: population std = {pop_std:.2f} | sample std = {sample_std:.2f}")

print("\n--- Standard Deviation: player_stats ---")
for col in ['goals', 'assists', 'minutes_played', 'yellow_cards']:
    pop_std = np.std(player_stats[col])
    sample_std = player_stats[col].std()
    print(f"{col}: population std = {pop_std:.2f} | sample std = {sample_std:.2f}")

print("\n--- Standard Deviation: teams (fifa_ranking_pre_tournament, elo_rating) ---")
print(teams[['fifa_ranking_pre_tournament', 'elo_rating']].std())     # pandas sample std, all columns at once

print("\n--- Standard Deviation: matches (home_score, away_score, home_xg, away_xg) ---")
played_matches = matches[matches['is_played']]
print(played_matches[['home_score', 'away_score', 'home_xg', 'away_xg']].std())

# -------------------------------------------------------------
# STEP 6C : COVARIANCE - NumPy & Pandas
# -------------------------------------------------------------
# Covariance shows the DIRECTION of a linear relationship between two variables
# (positive = move together, negative = move oppositely). Unlike correlation,
# it is NOT normalized, so its magnitude depends on the units/scale of the variables.

# Pairwise covariance (single pair) with NumPy -> np.cov returns a 2x2 matrix:
# [[var(x), cov(x,y)], [cov(x,y), var(y)]]
cov_matrix_np = np.cov(squads_and_players['market_value_eur'], squads_and_players['caps'])
print("\nNumPy covariance matrix (market_value_eur vs caps):\n", cov_matrix_np)
print("Covariance (market_value_eur, caps):", cov_matrix_np[0, 1])

# Full covariance matrix across several numeric columns (Pandas - easier for many columns)
cov_squads = squads_and_players[['market_value_eur', 'caps', 'height_cm', 'goals']].cov()
print("\nCovariance matrix - squads_and_players (market_value_eur, caps, height_cm, goals):\n", cov_squads)

cov_player_stats = player_stats[['goals', 'assists', 'shots_on_target', 'minutes_played',
                                   'yellow_cards', 'red_cards']].cov()
print("\nCovariance matrix - player_stats:\n", cov_player_stats)

cov_matches = played_matches[['home_score', 'away_score', 'home_xg', 'away_xg']].cov()
print("\nCovariance matrix - matches (scores & xG):\n", cov_matches)

# Covariance vs Correlation side-by-side for one pair (to show the relationship between them)
x, y = squads_and_players['market_value_eur'], squads_and_players['height_cm']
cov_xy = np.cov(x, y)[0, 1]
corr_xy = np.corrcoef(x, y)[0, 1]
print(f"\nmarket_value_eur vs height_cm -> covariance: {cov_xy:.2f} | correlation: {corr_xy:.4f}")
print("(Correlation = Covariance / (std_x * std_y) — this is why correlation is scale-free but covariance isn't.)")

# -------------------------------------------------------------
# STEP 7 : UNIQUE VALUES / CATEGORICAL EXPLORATION
# -------------------------------------------------------------
print("\nConfederations:\n", teams['confederation'].value_counts())
print("\nPositions:\n", squads_and_players['position'].value_counts())
print("\nTournament stages:\n", tournament_stages)
print("\nTop 10 venues by capacity:\n", venues.sort_values('capacity', ascending=False)[['stadium_name','city','capacity']].head(10))

# -------------------------------------------------------------
# STEP 8 : MERGING / JOINING TABLES (core relational EDA)
# -------------------------------------------------------------
# Matches + team names (join twice: home and away)
matches_full = matches.merge(teams[['team_id', 'team_name']], left_on='home_team_id',
                              right_on='team_id', how='left').rename(columns={'team_name': 'home_team_name'})
matches_full = matches_full.merge(teams[['team_id', 'team_name']], left_on='away_team_id',
                                    right_on='team_id', how='left', suffixes=('', '_away')).rename(
                                    columns={'team_name': 'away_team_name'})
matches_full = matches_full.drop(columns=[c for c in matches_full.columns if c.startswith('team_id')])

# Add venue info
matches_full = matches_full.merge(venues[['venue_id', 'stadium_name', 'city', 'country']],
                                    on='venue_id', how='left')

# Add referee info
matches_full = matches_full.merge(referees[['referee_id', 'name']], on='referee_id', how='left')
matches_full = matches_full.rename(columns={'name': 'referee_name'})

print("\nMerged matches_full shape:", matches_full.shape)
print(matches_full[['date', 'home_team_name', 'away_team_name', 'home_score',
                    'away_score', 'stadium_name', 'referee_name']].head())

# Player stats + team name
player_stats_full = player_stats.merge(teams[['team_id', 'team_name']], on='team_id', how='left')
print("\nTop 10 goal scorers:\n",
      player_stats_full.sort_values('goals', ascending=False)[['player_name', 'team_name', 'goals', 'assists']].head(10))

# Squad + player stats combined (player-level master table)
player_master = squads_and_players.merge(
    player_stats[['player_id', 'matches_played', 'minutes_played', 'average_rating', 'yellow_cards', 'red_cards']],
    on='player_id', how='left', suffixes=('', '_stats')
)
print("\nPlayer master table shape:", player_master.shape)

# -------------------------------------------------------------
# STEP 9 : GROUPBY / AGGREGATION ACROSS MERGED DATA
# -------------------------------------------------------------
goals_by_team = player_stats_full.groupby('team_name')['goals'].sum().sort_values(ascending=False)
print("\nTotal goals by team (top 10):\n", goals_by_team.head(10))

avg_rating_by_position = player_master.groupby('position')['average_rating'].mean().sort_values(ascending=False)
print("\nAverage rating by position:\n", avg_rating_by_position)

matches_per_venue = matches_full.groupby('stadium_name').size().sort_values(ascending=False)
print("\nMatches hosted per venue:\n", matches_per_venue.head(10))

cards_by_referee = referees.sort_values('avg_cards_per_game', ascending=False)
print("\nReferees by avg cards per game:\n", cards_by_referee[['name', 'country', 'avg_cards_per_game']].head(10))

home_away_goal_avg = matches[matches['is_played']][['home_score', 'away_score']].mean()
print("\nAverage home vs away goals:\n", home_away_goal_avg)

# -------------------------------------------------------------
# STEP 10 : EVENT-LEVEL ANALYSIS (match_events)
# -------------------------------------------------------------
print("\nEvent type distribution:\n", match_events['event_type'].value_counts())

events_per_match = match_events.groupby('match_id').size().sort_values(ascending=False)
print("\nMatches with most events:\n", events_per_match.head(5))

goal_events = match_events[match_events['event_type'].str.contains('goal', case=False, na=False)]
print("\nTotal goal-type events:", len(goal_events))

# -------------------------------------------------------------
# STEP 11 : LINEUP ANALYSIS
# -------------------------------------------------------------
starting_xi_count = match_lineups.groupby('match_id')['is_starting_xi'].sum()
print("\nStarting XI count per match (should be ~22):\n", starting_xi_count.describe())

avg_minutes_by_position = match_lineups.merge(
    squads_and_players[['player_id', 'position']], on='player_id', how='left'
).groupby('position')['minutes_played'].mean()
print("\nAvg minutes played by position:\n", avg_minutes_by_position)

# -------------------------------------------------------------
# STEP 12 : OUTLIER DETECTION (IQR + Z-SCORE)
# -------------------------------------------------------------
def detect_outliers_iqr(dataframe, column):
    Q1, Q3 = dataframe[column].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    return dataframe[(dataframe[column] < lower) | (dataframe[column] > upper)]

outliers_market_value = detect_outliers_iqr(squads_and_players, 'market_value_eur')
print("\nOutliers in market_value_eur:", outliers_market_value.shape[0])

if player_stats['average_rating'].std() > 0:
    z = (player_stats['average_rating'] - player_stats['average_rating'].mean()) / player_stats['average_rating'].std()
    outliers_rating = player_stats[np.abs(z) > 3]
    print("Outliers via z-score in average_rating:", outliers_rating.shape[0])
else:
    print("Skipping z-score outliers for average_rating: column has no variation (likely all zero/empty in source data).")

# -------------------------------------------------------------
# STEP 13 : FEATURE ENGINEERING
# -------------------------------------------------------------
squads_and_players['age'] = (pd.Timestamp('2026-06-01') - squads_and_players['date_of_birth']).dt.days // 365
player_master['goal_involvement'] = player_master['average_rating'].notnull().astype(int)  # placeholder flag example

matches['total_goals'] = matches['home_score'] + matches['away_score']
matches['goal_difference'] = (matches['home_score'] - matches['away_score']).abs()
matches['margin_type'] = np.where(matches['goal_difference'] >= 3, 'Blowout',
                            np.where(matches['goal_difference'] == 0, 'Draw', 'Close'))

# -------------------------------------------------------------
# STEP 14 : NUMPY ARRAY-LEVEL OPERATIONS
# -------------------------------------------------------------
ratings_array = player_stats['average_rating'].to_numpy()
if player_stats['average_rating'].sum() != 0:
    print("\nCumulative sum of first 10 ratings:", np.cumsum(ratings_array)[:10])
    print("Players with rating > 8:", np.sum(ratings_array > 8))
else:
    print("\nNote: average_rating is empty/zero for all players in the source data — skipping rating-based stats.")
print("Any red cards issued:", np.any(player_stats['red_cards'].to_numpy() > 0))
print("All players played >= 0 minutes:", np.all(player_stats['minutes_played'].to_numpy() >= 0))

# -------------------------------------------------------------
# STEP 15 : CONSISTENCY / REFERENTIAL INTEGRITY CHECKS
# -------------------------------------------------------------
orphan_players = squads_and_players[~squads_and_players['team_id'].isin(teams['team_id'])]
print("\nPlayers with team_id not in teams table:", len(orphan_players))

orphan_matches = matches[~matches['venue_id'].isin(venues['venue_id'])]
print("Matches with venue_id not in venues table:", len(orphan_matches))

# -------------------------------------------------------------
# STEP 16 : EXPORT CLEANED / MERGED DATA
# -------------------------------------------------------------
matches_full.to_csv('matches_full_cleaned.csv', index=False)
player_master.to_csv('player_master_cleaned.csv', index=False)
print("\nExported: matches_full_cleaned.csv, player_master_cleaned.csv")

# -------------------------------------------------------------
# STEP 17 : SUMMARY REPORT
# -------------------------------------------------------------
with open('eda_summary_report.txt', 'w') as f:
    f.write("FIFA WORLD CUP 2026 - EDA SUMMARY\n")
    f.write("="*50 + "\n")
    for name, df in tables.items():
        f.write(f"{name}: shape={df.shape}\n")
    f.write(f"\nTotal matches played: {matches['is_played'].sum()} / {len(matches)}\n")
    f.write(f"Top scoring team: {goals_by_team.index[0]} ({goals_by_team.iloc[0]} goals)\n")
    if player_stats['average_rating'].sum() == 0:
        f.write("Note: 'average_rating' (and shots/shots_on_target) are empty in the source data for all players.\n")
    else:
        f.write(f"Top rated position: {avg_rating_by_position.index[0]}\n")

print("\nFull relational EDA complete. Report saved to 'eda_summary_report.txt'.")
```

```python
# =============================================================
# FIFA WORLD CUP 2026 - COMPLETE VISUALIZATION EDA
# Seaborn + Matplotlib (all major chart types) + KPI Dashboard + Geo Maps
# =============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

# -------------------------------------------------------------
# STEP 0 : LOAD DATA
# -------------------------------------------------------------
matches   = pd.read_csv('matches.csv')
teams     = pd.read_csv('teams.csv')
venues    = pd.read_csv('venues.csv')
referees  = pd.read_csv('referees.csv')
players   = pd.read_csv('squads_and_players.csv')
player_stats = pd.read_csv('player_stats.csv')

# Merge for richer plots (team names, venue info on matches)
matches_full = (matches
    .merge(teams[['team_id', 'team_name']], left_on='home_team_id', right_on='team_id', how='left')
    .rename(columns={'team_name': 'home_team_name'}).drop(columns='team_id')
    .merge(teams[['team_id', 'team_name']], left_on='away_team_id', right_on='team_id', how='left')
    .rename(columns={'team_name': 'away_team_name'}).drop(columns='team_id')
    .merge(venues, on='venue_id', how='left')
    .merge(referees[['referee_id', 'name']], on='referee_id', how='left')
    .rename(columns={'name': 'referee_name'}))

matches_full['is_played'] = matches_full['home_score'].notnull()
played = matches_full[matches_full['is_played']].copy()
played['total_goals'] = played['home_score'] + played['away_score']

player_stats_full = player_stats.merge(teams[['team_id', 'team_name']], on='team_id', how='left')
goals_by_team = player_stats_full.groupby('team_name')['goals'].sum().sort_values(ascending=False)


# =================================================================
# PART 1 : DISTRIBUTION CHARTS
# =================================================================

# 1. Histogram
plt.figure()
sns.histplot(players['market_value_eur'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Player Market Value (EUR)')
plt.savefig('01_hist_market_value.png'); plt.close()

# 2. KDE plot
plt.figure()
sns.kdeplot(players['height_cm'], fill=True, color='green')
plt.title('Player Height Density (KDE)')
plt.savefig('02_kde_height.png'); plt.close()

# 3. Rug + histogram combo
plt.figure()
sns.histplot(played['total_goals'], bins=10, color='orange')
sns.rugplot(played['total_goals'], color='black')
plt.title('Total Goals per Match (with Rug)')
plt.savefig('03_hist_rug_goals.png'); plt.close()

# 4. ECDF plot (cumulative distribution)
plt.figure()
sns.ecdfplot(players['caps'])
plt.title('ECDF - Player International Caps')
plt.savefig('04_ecdf_caps.png'); plt.close()


# =================================================================
# PART 2 : CATEGORICAL / COMPARISON CHARTS
# =================================================================

# 5. Countplot
plt.figure()
sns.countplot(y='position', data=players, order=players['position'].value_counts().index,
              hue='position', palette='viridis', legend=False)
plt.title('Player Count by Position')
plt.savefig('05_count_position.png'); plt.close()

# 6. Bar chart (aggregated)
plt.figure()
top10_goals = goals_by_team.head(10)
sns.barplot(x=top10_goals.values, y=top10_goals.index, hue=top10_goals.index,
            palette='rocket', legend=False)
plt.title('Top 10 Teams by Total Goals')
plt.xlabel('Goals')
plt.savefig('06_bar_top_scoring_teams.png'); plt.close()

# 7. Horizontal stacked bar (wins/draws/losses style using match results)
home_wins = (played['home_score'] > played['away_score']).sum()
away_wins = (played['away_score'] > played['home_score']).sum()
draws     = (played['home_score'] == played['away_score']).sum()
plt.figure()
plt.bar(['Result'], [home_wins], color='#4C72B0', label='Home Wins')
plt.bar(['Result'], [away_wins], bottom=[home_wins], color='#DD8452', label='Away Wins')
plt.bar(['Result'], [draws], bottom=[home_wins + away_wins], color='#55A868', label='Draws')
plt.title('Match Outcome Breakdown (Stacked Bar)')
plt.legend()
plt.savefig('07_stacked_bar_outcomes.png'); plt.close()

# 8. Boxplot
plt.figure()
sns.boxplot(x='position', y='market_value_eur', data=players, hue='position',
            palette='Set2', legend=False)
plt.title('Market Value by Position')
plt.xticks(rotation=20)
plt.savefig('08_box_marketvalue_position.png'); plt.close()

# 9. Violin plot
plt.figure()
sns.violinplot(x='position', y='height_cm', data=players, hue='position',
               palette='muted', legend=False)
plt.title('Height Distribution by Position (Violin)')
plt.savefig('09_violin_height_position.png'); plt.close()

# 10. Strip plot
plt.figure()
sns.stripplot(x='position', y='caps', data=players, hue='position',
              palette='cool', alpha=0.5, legend=False)
plt.title('Caps by Position (Strip Plot)')
plt.savefig('10_strip_caps_position.png'); plt.close()

# 11. Swarm plot (sampled for readability on large data)
plt.figure()
sample_players = players.sample(min(300, len(players)), random_state=1)
sns.swarmplot(x='position', y='goals', data=sample_players, hue='position',
              palette='Set3', size=3, legend=False)
plt.title('Goals by Position (Swarm Plot, sampled)')
plt.savefig('11_swarm_goals_position.png'); plt.close()

# 12. Boxen plot (letter-value plot, good for larger data)
plt.figure()
sns.boxenplot(x='position', y='market_value_eur', data=players, hue='position',
              palette='crest', legend=False)
plt.title('Market Value by Position (Boxen Plot)')
plt.savefig('12_boxen_marketvalue_position.png'); plt.close()

# 13. Pie chart
plt.figure()
confed_counts = teams['confederation'].value_counts()
plt.pie(confed_counts, labels=confed_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Team Share by Confederation')
plt.savefig('13_pie_confederation.png'); plt.close()

# 14. Donut chart
plt.figure()
plt.pie(confed_counts, labels=confed_counts.index, autopct='%1.1f%%',
        colors=sns.color_palette('pastel'), wedgeprops=dict(width=0.4))
plt.title('Team Share by Confederation (Donut)')
plt.savefig('14_donut_confederation.png'); plt.close()


# =================================================================
# PART 3 : RELATIONSHIP CHARTS
# =================================================================

# 15. Scatter plot
plt.figure()
sns.scatterplot(x='market_value_eur', y='caps', data=players, hue='position', alpha=0.6)
plt.title('Market Value vs Caps')
plt.savefig('15_scatter_marketvalue_caps.png'); plt.close()

# 16. Regression plot (scatter + trend line)
plt.figure()
sns.regplot(x='height_cm', y='market_value_eur', data=players,
            scatter_kws={'alpha': 0.3}, line_kws={'color': 'red'})
plt.title('Height vs Market Value (with Regression Line)')
plt.savefig('16_regplot_height_marketvalue.png'); plt.close()

# 17. Hexbin joint plot
sns.jointplot(x='home_score', y='away_score', data=played, kind='hex', color='darkorange')
plt.savefig('17_jointplot_scores.png'); plt.close()

# 18. Pairplot
subset = players[['market_value_eur', 'caps', 'height_cm', 'goals']].sample(min(500, len(players)), random_state=1)
sns.pairplot(subset, diag_kind='kde')
plt.savefig('18_pairplot_players.png'); plt.close()

# 19. Correlation heatmap
plt.figure(figsize=(8, 6))
corr = players[['market_value_eur', 'caps', 'height_cm', 'goals']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap - Player Attributes')
plt.savefig('19_heatmap_correlation.png'); plt.close()

# 20. Clustermap (hierarchical clustering + heatmap)
sns.clustermap(corr, annot=True, cmap='vlag', figsize=(7, 7))
plt.savefig('20_clustermap_correlation.png'); plt.close()

# 21. Pivot heatmap (team x position player count)
pivot = players.merge(teams[['team_id', 'team_name']], on='team_id', how='left')
pivot_table = pd.pivot_table(pivot, index='team_name', columns='position',
                              values='player_id', aggfunc='count', fill_value=0)
plt.figure(figsize=(10, 14))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=False)
plt.title('Player Count: Team x Position')
plt.savefig('21_heatmap_team_position.png'); plt.close()


# =================================================================
# PART 4 : TIME / TREND CHARTS
# =================================================================

matches_full['date'] = pd.to_datetime(matches_full['date'], errors='coerce')
played['date'] = pd.to_datetime(played['date'], errors='coerce')
goals_by_date = played.groupby(played['date'].dt.date)['total_goals'].sum()

# 22. Line plot
plt.figure()
goals_by_date.plot(kind='line', marker='o', color='teal')
plt.title('Total Goals per Match Day')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('22_line_goals_by_date.png'); plt.close()

# 23. Area chart
plt.figure()
goals_by_date.plot(kind='area', alpha=0.4, color='steelblue')
plt.title('Total Goals per Match Day (Area Chart)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('23_area_goals_by_date.png'); plt.close()

# 24. Cumulative line chart
plt.figure()
goals_by_date.cumsum().plot(kind='line', color='purple', marker='o')
plt.title('Cumulative Goals Across Tournament')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('24_line_cumulative_goals.png'); plt.close()


# =================================================================
# PART 5 : FACETED / MULTI-PANEL CHARTS
# =================================================================

# 25. FacetGrid
g = sns.FacetGrid(players, col='position', col_wrap=3, height=3.5)
g.map(sns.histplot, 'market_value_eur', bins=15, color='steelblue')
g.savefig('25_facetgrid_marketvalue_by_position.png'); plt.close()

# 26. Multi-subplot dashboard (2x2)
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
sns.histplot(players['market_value_eur'], bins=20, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Market Value Distribution')
sns.boxplot(x='position', y='height_cm', data=players, hue='position',
            ax=axes[0, 1], palette='Set3', legend=False)
axes[0, 1].set_title('Height by Position')
sns.scatterplot(x='caps', y='market_value_eur', data=players, ax=axes[1, 0], alpha=0.5)
axes[1, 0].set_title('Caps vs Market Value')
sns.barplot(x=goals_by_team.head(5).values, y=goals_by_team.head(5).index,
            hue=goals_by_team.head(5).index, ax=axes[1, 1], palette='mako', legend=False)
axes[1, 1].set_title('Top 5 Scoring Teams')
plt.tight_layout()
plt.savefig('26_dashboard_subplots.png'); plt.close()


# =================================================================
# PART 6 : KPI DASHBOARD (text/number cards using Matplotlib)
# =================================================================

total_matches      = len(matches_full)
matches_played      = played.shape[0]
total_goals_scored  = played['total_goals'].sum()
avg_goals_per_match = played['total_goals'].mean()
total_teams          = teams.shape[0]
total_players        = players.shape[0]
top_team, top_goals   = goals_by_team.index[0], goals_by_team.iloc[0]
avg_market_value     = players['market_value_eur'].mean()

kpis = [
    ("Total Matches", f"{total_matches}"),
    ("Matches Played", f"{matches_played}"),
    ("Total Goals", f"{int(total_goals_scored)}"),
    ("Avg Goals / Match", f"{avg_goals_per_match:.2f}"),
    ("Total Teams", f"{total_teams}"),
    ("Total Players", f"{total_players}"),
    ("Top Scoring Team", f"{top_team} ({top_goals})"),
    ("Avg Market Value", f"€{avg_market_value/1e6:.1f}M"),
]

fig = plt.figure(figsize=(14, 6))
gs = gridspec.GridSpec(2, 4, figure=fig)
for i, (label, value) in enumerate(kpis):
    ax = fig.add_subplot(gs[i // 4, i % 4])
    ax.axis('off')
    ax.text(0.5, 0.6, value, ha='center', va='center', fontsize=22, fontweight='bold', color='#2c3e50')
    ax.text(0.5, 0.2, label, ha='center', va='center', fontsize=11, color='#7f8c8d')
    ax.set_facecolor('#f7f9fa')
    for spine in ax.spines.values():
        spine.set_visible(True)
        spine.set_edgecolor('#dcdcdc')
fig.suptitle('FIFA World Cup 2026 - Key Performance Indicators', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('27_kpi_dashboard.png'); plt.close()


# =================================================================
# PART 7 : GEO / MAP VISUALIZATIONS
# =================================================================

# 28. Basic Matplotlib geo-scatter (no internet/tiles needed - always works)
plt.figure(figsize=(10, 6))
sizes = venues['capacity'] / 300
plt.scatter(venues['longitude'], venues['latitude'], s=sizes, c=venues['elevation_meters'],
            cmap='viridis', alpha=0.8, edgecolors='black')
for _, row in venues.iterrows():
    plt.annotate(row['city'], (row['longitude'], row['latitude']), fontsize=7, alpha=0.7)
plt.colorbar(label='Elevation (m)')
plt.xlabel('Longitude'); plt.ylabel('Latitude')
plt.title('World Cup Venues - Location, Capacity (size) & Elevation (color)')
plt.savefig('28_geo_scatter_venues.png'); plt.close()

# 29. Plotly interactive geo map (scatter_geo) — needs: !pip install plotly (pre-installed on Colab)
try:
    import plotly.express as px
    fig_geo = px.scatter_geo(
        venues, lat='latitude', lon='longitude', text='city',
        size='capacity', color='elevation_meters',
        color_continuous_scale='viridis',
        scope='north america',
        title='World Cup 2026 Venues (Interactive Geo Map)'
    )
    fig_geo.update_layout(geo=dict(showland=True, landcolor='rgb(240,240,240)'))
    fig_geo.write_html('29_plotly_geo_venues.html')
    fig_geo.show()   # renders inline in Colab
except ImportError:
    print("Plotly not installed. Run: !pip install plotly")

# 30. Folium interactive map with markers (needs: !pip install folium, pre-installed on Colab)
try:
    import folium
    m = folium.Map(location=[venues['latitude'].mean(), venues['longitude'].mean()],
                    zoom_start=4, tiles='CartoDB positron')
    for _, row in venues.iterrows():
        folium.CircleMarker(
            location=[row['latitude'], row['longitude']],
            radius=max(row['capacity'] / 8000, 4),
            popup=f"{row['stadium_name']}<br>{row['city']}, {row['country']}<br>Capacity: {row['capacity']:,}",
            color='crimson', fill=True, fill_color='crimson', fill_opacity=0.6
        ).add_to(m)
    m.save('30_folium_map_venues.html')
    print("Folium map saved to 30_folium_map_venues.html — open this file or display `m` directly in a notebook cell.")
except ImportError:
    print("Folium not installed. Run: !pip install folium")

print("\nAll visualizations (charts + KPI dashboard + geo maps) generated successfully.")
```
